# Traveling Salesman

```python
import numpy as np
from pulp import *
import matplotlib.pyplot as plt
```

```python
N = 50
X = 100*np.random.rand(N,2)
nejlevejsi=np.argmin(X[:,0])
X[0,:],X[nejlevejsi,:]=X[nejlevejsi,:], X[0,:]

C = np.zeros((N,N))
for i in range(N):
    for j in range(N):
        C[i,j] = np.sqrt((X[i,0]-X[j,0])**2+(X[i,1]-X[j,1])**2)
plt.scatter(X[1:,0],X[1:,1])
plt.scatter(X[0,0],X[0,1], color="red")
model = LpProblem("Cesta", LpMinimize)
```
